```python
result['feature_1'] = result.groupby('user_id', as_index=False)['ride_cost'].transform(lambda x: x.quantile(0.75))
result['feature_2'] = result.groupby('user_id', as_index=False)['rating'].transform(lambda x: (x.max() - x.min()) / x.mean())
result['feature_3'] = result.groupby('car_id', as_index=False)['ride_duration'].transform(lambda x: x[x > 40].count())
result['feature_4'] = result.groupby('car_id', as_index=False)['ride_duration'].transform(lambda x: x[x > 40].sum())

result = result.fillna(0)
result['feature_3'] = result['feature_3'].astype('int64')
result['feature_4'] = result['feature_4'].astype('int64')
```

```python
import pandas as pd

df = pd.read_csv('https://stepik.org/media/attachments/lesson/779916/3_1_7.csv')

# В ans записан правильный ответ
ans = pd.read_csv('https://stepik.org/media/attachments/lesson/779916/3_1_7_ans.csv')
```

```python
df.head()
```

```python
pivot_table1 = df[['user_id', 'car_type', 'car_id']].pivot_table(
    index='user_id',
    columns='car_type',
    values='car_id',
    aggfunc='count'
).fillna(0)
pivot_table1.columns = [f'{i}_count' for i in pivot_table1.columns]
pivot_table1 = pivot_table1.reset_index()
```

```python
pivot_table1.head()
```

```python
table2 = df[['user_id', 'distance', 'ride_cost']].groupby('user_id', as_index=False).agg(['min', 'max', 'mean'])
table2.columns = [f'{i[0]}_{i[1]}' for i in table2.columns]
table2 = table2.reset_index()
```

```python
table2.head()
```

```python
table3 = df[['user_id', 'car_id']].groupby('user_id', as_index=False).agg('nunique').rename(columns={'car_id': 'car_id_nunique'})
```

```python
table3.head()
```

```python
result = pd.merge(pivot_table1, pd.merge(table2, table3, on='user_id'), on='user_id')
```

```python
result.head()
```

```python
result['crazy_driver'] = df[['user_id', 'speed_max']].groupby('user_id', as_index=False).agg('max')['speed_max'] > 100
```

```python
result.columns = ['user_id',
                  'economy_count',
                  'standart_count',
                  'distance_min',
                  'ride_cost_min',
                  'distance_max',
                  'ride_cost_max',
                  'distance_mean',
                  'ride_cost_mean',
                  'car_id_nunique',
                  'crazy_driver']
```

```python
result['ride_cost_mean'] = result['ride_cost_mean'].astype('int64')
result['ride_cost_min'] = result['ride_cost_min'].astype('int64')
result['ride_cost_max'] = result['ride_cost_max'].astype('int64')
result['crazy_driver'] = result['crazy_driver'].astype('int64')
```

```python
result = result[['user_id',
                  'economy_count',
                  'standart_count',
                  'distance_min',
                  'ride_cost_min',
                  'distance_max',
                  'ride_cost_max',
                  'distance_mean',
                  'ride_cost_mean',
                  'car_id_nunique',
                  'crazy_driver']]
```
